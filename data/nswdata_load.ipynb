```python
# 提取ID和经纬度
id_coords = data_location[['ID', 'Latitude', 'Longitude']]

# 将ID设为索引，方便查找
id_coords.set_index('ID', inplace=True)
from itertools import combinations

# 获取所有ID，排除日期列
id_list = data_sensor.columns.tolist()
id_list.remove('date')

# 生成所有ID对
id_pairs = list(combinations(id_list, 2))
import requests

def get_driving_distance(coord1, coord2):
    url = f'http://localhost:5000/route/v1/driving/{coord1[1]},{coord1[0]};{coord2[1]},{coord2[0]}?overview=false'
    response = requests.get(url)
    data = response.json()
    if data['code'] == 'Ok':
        # 距离以米为单位
        distance = data['routes'][0]['distance']
        return distance
    else:
        return None
# 创建一个空的结果列表
results = []

for id1, id2 in id_pairs:
    # 获取ID的坐标
    try:
        coord1 = id_coords.loc[int(id1), ['Latitude', 'Longitude']]
        coord2 = id_coords.loc[int(id2), ['Latitude', 'Longitude']]
    except KeyError:
        # 如果ID在id_coords中不存在，跳过
        continue

    # 计算行驶距离
    distance = get_driving_distance(coord1, coord2)
    if distance is not None:
        results.append({'ID1': id1, 'ID2': id2, 'Distance_m': distance})
df_results = pd.DataFrame(results)
df_results
```

```python
final_df = pd.DataFrame()
for file in csv_files:
    # 从文件名中提取数字
    filename = os.path.basename(file)
    match = re.search(r'(\d+)', filename)
    if match:
        column_name = match.group(1)
    else:
        continue  # 如果文件名中没有数字，跳过这个文件
    
    # 读取CSV文件
    df = pd.read_csv(file)
                                
    df.drop_duplicates(inplace=True)
    # 筛选包含“All Vehicles”的行
    df = df[df['classification_seq'] == 'All Vehicles']
    
    # 创建一个空的DataFrame来保存转换后的数据
    transformed_df = pd.DataFrame()
    
    # 遍历每一行数据
    for _, row in df.iterrows():
        # 对于每一行数据，我们需要将24个小时的数据转换为24行
        for hour in range(24):
            hour_column = f'hour_{hour:02}'
            if hour_column in row:
                # 创建一个新行
                new_row = {
                    'date': pd.Timestamp(f"{row['date']} {hour:02}:00:00"),
                    'direction': row['cardinal_direction_seq'],
                    column_name: row[hour_column]  # 使用文件名中的数字作为列名
                }
                # 添加到转换DataFrame
                transformed_df = transformed_df.append(new_row, ignore_index=True)
    
    # 只保留2013年1月2日及以后的数据
    transformed_df = transformed_df[transformed_df['date'] >= '2013-01-02']
    
    # 合并到最终的DataFrame
    if final_df.empty:
        final_df = transformed_df
    else:
        final_df = pd.concat([final_df, transformed_df], ignore_index=True)

# 聚合数据，按日期进行分组，并且聚合所有方向的数据
final_df = final_df.groupby('date').agg(lambda x: x.dropna().unique().sum())
```
